### **Рассмотрим значения кластеров и их статистику**

```python
columns = [
    'region',
    'min_living_cost',
    'per_capita_cash_income',
    'formal_wage_paid',
    'gross_product',
    'retail_turnover',
    'regional_prod',
    'disease_alco',
    'disease_drug',
    'registered_crimes',
    'ophans',
    'cluster',
    'unemployment',
    'poverty_percent'
]
# создаем выборку основных признаков
df_regions_main = df_regions[columns].copy().set_index('region')

# стандартизуем признаки на базовый показатель min_living_cost
df_regions_main['per_capita_cash_income_stand'] = df_regions_main['per_capita_cash_income'] / df_regions_main['min_living_cost']
df_regions_main['formal_wage_paid_stand'] = df_regions_main['formal_wage_paid'] / df_regions_main['min_living_cost']
df_regions_main['retail_turnover_stand'] = df_regions_main['retail_turnover'] / (df_regions_main['min_living_cost'] * 12)

# удаляем исходные признаки
df_regions_main.drop([
    'per_capita_cash_income',
    'formal_wage_paid',
    'retail_turnover',
    'min_living_cost'
], axis=1, inplace=True)

# логарифмируем призанки
log_cols = [
    'gross_product',
    'disease_drug',
    'unemployment',
    'regional_prod'
]

for col in log_cols:
    if col == 'disease_drug':
        df_regions_main['log_' + col] = np.log(df_regions_main[col] + 1)
        continue
    # другие признаки - просто логарифмирование
    df_regions_main['log_' + col] = np.log(df_regions_main[col])
# удаляем исходные признаки    
df_regions_main.drop(log_cols, axis=1, inplace=True)

# для лучшего визуального восприятия упорядочиваем колонки
columns = [
    'poverty_percent', 'log_unemployment', 'disease_alco', 
    'registered_crimes', 'ophans', 'log_disease_drug', 'per_capita_cash_income_stand',
    'formal_wage_paid_stand','retail_turnover_stand', 'log_gross_product', 'log_regional_prod', 'cluster'
]
df_regions_main = df_regions_main[columns].copy()

# группируем данные по кластерам
df_grouped = df_regions_main.groupby('cluster').median()
display(df_grouped)
```

Строим визуализацию.

```python

def plot_cluster_profiles_1x4_final(medians, n_clusters):
    # 1. Нормализация [0,1] 
    scaler = MinMaxScaler()
    norm_df = pd.DataFrame(
        scaler.fit_transform(medians),
        columns=medians.columns,
        index=medians.index
    )
    features = medians.columns.tolist()

    # 2. Сетка
    fig = make_subplots(
        rows=1,
        cols=n_clusters,
        specs=[[{'type': 'polar'}] * n_clusters],
        horizontal_spacing=0.14           # отступ между графиками
    )

    # 3. Добавляем каждый кластер (стандартные цвета Plotly)
    for i in range(n_clusters):
        fig.add_trace(
            go.Scatterpolar(
                r=norm_df.iloc[i].values,
                theta=features,
                fill='toself',
                name=f'Кластер {i}',
                showlegend=True  # легенда включена
            ),
            row=1, col=i + 1
        )

        # Единые оси
        fig.update_polars(
            radialaxis=dict(
                visible=True,
                range=[0, 1],
                showticklabels=False,
                ticks=''
            ),
            angularaxis=dict(
                rotation=90,
                direction='clockwise',
                tickfont=dict(size=10)
            ),
            row=1, col=i + 1
        )

    # 4. Общие настройки: отступы, размер, легенда 
    fig.update_layout(
        title_text="Профили кластеров",
        title_x=0.5,
        title_font=dict(size=16),
        height=600,
        width=1800,
        # Легенда: сверху, по центру
        legend=dict(
            orientation="h",
            yanchor="bottom",
            y=1.02,
            xanchor="center",
            x=0.5,
            font=dict(size=12)
        ),
        font=dict(size=11),
    )

    fig.show()


plot_cluster_profiles_1x4_final(df_grouped, n_clusters=4)
```

```python
cols = df_regions_main.drop('cluster', axis=1).columns

# 1. Создаём сетку 3×4 
fig, ax = plt.subplots(3, 4, figsize=(25, 11))
fig.suptitle('Распределения характеристик по кластерам', 
             fontsize=20, fontweight='bold', y=1.02)

# 2. Заполняем графики 
i = 0
for row in range(3):
    for col in range(4):
        if i < len(cols):
            feature = cols[i]
            sns.boxplot(
                data=df_regions_main,
                x=feature,
                y='cluster',
                orient='h',
                ax=ax[row, col],
                palette='Set2',  
                linewidth=1.5,
                fliersize=4
            )
            ax[row, col].set_title(feature.replace('_', ' ').title(), fontsize=14, pad=10)
            ax[row, col].set_ylabel('')  # убираем подпись оси Y
            ax[row, col].set_xlabel('')  # убираем подпись оси X
            ax[row, col].grid(True, axis='x', linestyle='--', alpha=0.6)
        else:
            ax[row, col].axis('off')  # скрываем пустые ячейки
        i += 1

plt.tight_layout()  
plt.show()
```

### **Cтатистическая проверка различий между кластерами**

Для проверки статистической значимости различий между четырьмя кластерами по 11 ключевым признакам использован **Kruskal-Wallis H-test** — непараметрический аналог однофакторного дисперсионного анализа.

Гипотезы:
- **Нулевая гипотеза (H₀)**:  
  Значения всех четырёх кластеров по данному признаку **равны** (распределения не различаются).
- **Альтернативная гипотеза (H₁)**:  
  Значения **хотя бы двух кластеров** различаются (существуют статистически значимые различия).

Уровень значимости: **α = 0.05**

```python
def kruskal_test(df, col):
    """Kruskal-Wallis + вывод решения по H₀/H₁"""
    v0 = df[df['cluster'] == 0][col]
    v1 = df[df['cluster'] == 1][col]
    v2 = df[df['cluster'] == 2][col]
    v3 = df[df['cluster'] == 3][col]
    _, p = kruskal(v0, v1, v2, v3)
    
    print(f"{col}:")
    decision_hypothesis(p)
    return p

for col in cols:
    kruskal_test(df_regions_main, col)
```

**Вывод:**
-   Кластеры статистически различаются по основным признакам

### **Влияет ли число детей, пенсионеров и других социально уязвимых групп на уровень бедности в регионе**

```python
social_cols = ['poverty_children', 'poverty_pensioners', 'poverty_labouring']

# кол-во малообеспеченного населения
df_regions['poverty_cnt'] = (df_regions.population * (df_regions.poverty_percent / 100)).round()

# кол-во малообеспеченных детей, пенсионеров, трудоспособных людей (абсолютные числа)
df_regions['poverty_children_cnt'] = (df_regions['poverty_cnt'] * (df_regions['poverty_children'] / 100)).round()
df_regions['poverty_pensioners_cnt'] = (df_regions['poverty_cnt'] * (df_regions['poverty_pensioners'] / 100)).round()
df_regions['poverty_labouring_cnt'] = (df_regions['poverty_cnt'] * (df_regions['poverty_labouring'] / 100)).round()
```

```python
features = [
    'poverty_children_cnt', 'poverty_pensioners_cnt',
    'poverty_labouring_cnt', 'poverty_cnt', 'unemployment'
]

# построим коореляционную карту
sns.heatmap(df_regions[features].corr('spearman'), annot=True)
plt.title('Корреляция демографических признаков и уровня бедности')
plt.show()
```

Напоминаю что под уровнем бедности в данном этапе исследования подразумевается признак *poverty_percent*. Для ответа на вопрос построили тепловую карту корреляции, для поиска линейной зависисмости. При ответе на вопрос были использованы признаки демографических уязвимых групп населения таких как: 'poverty_children_cnt', 'poverty_pensioners_cnt', 'poverty_labouring_cnt'

По результатам построенной корреляционной карты была выявленна очень сильная корреляционная связь, что дает основания для принятия решения о их линейной взаимосвязи. 

> **Влияет ли число детей, пенсионеров и других социально уязвимых групп на уровень бедности в регионе?**

**Ответ**: 
-   Непременно влияет, эти показатели напрямую связанны с показателем бедности.  Сам уровень бедности из себя предсталвяет процент людей живущих за чертой бедности, то есть малообеспеченное население. Если говорить обобщенно, мы брали показатели количество малообеспеченых людей определенной демографической группы, если брать в целом количество детей, пенсионеров, трудящихся не включенных в группу *малообеспеченное население*, тогда корреляция будет далеко не такой сильной, она скорее будет слабой, так как напрямую уровень бедности не зависит от кол-ва населения. Понятное дело бывают частные случаи где связь может быть более сильной, но говоря в целом связь будет слабой.
