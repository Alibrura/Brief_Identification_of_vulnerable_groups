#### `Функция для извлечения данных и стандартизации названия регионов.`

```python
def process_region_data(dataframe, standard_names_dict, exclude_list=None, region_column='region'):
    """
    Обрабатывает данные по федеральным субъектам, стандартизируя названия регионов.
    
    Args:
        dataframe (pd.DataFrame): Входной DataFrame с данными.
        standard_names_dict (dict): Словарь соответствия нестандартных названий стандартным.
        exclude_list (list, optional): Список названий регионов для исключения.
        region_column (str, optional): Название столбца с регионами. По умолчанию 'region'.
    
    Returns:
        pd.DataFrame: Обработанный DataFrame с стандартизированными названиями регионов.
    """
    
    def standardize_region_name(region_name):
        """Преобразует название региона в стандартный вид."""
        if pd.isna(region_name):
            return None
        # Удаляем текст в скобках и лишние пробелы
        cleaned_name = re.sub(r'\s*\([^)]+\)', '', str(region_name)).strip()
        # Удаляем переносы строк
        cleaned_name = cleaned_name.replace('\n', ' ')
        # Также заменяем множественные пробелы на один
        cleaned_name = re.sub(r'\s+', ' ', cleaned_name)
        
        # Поиск совпадения с ключами словаря (игнорируя регистр)
        for pattern, standard_name in standard_names_dict.items():
            if re.search(pattern.lower(), cleaned_name.lower()):
                return standard_name
        return None
    
    # Копируем DataFrame для избежания изменения исходных данных
    df = dataframe.copy()
    
    # Исключаем регионы из списка исключений, если он задан
    if exclude_list is not None:
        df = df[~df[region_column].isin(exclude_list)]

    # Применяем стандартизацию названий
    df[region_column] = df[region_column].apply(standardize_region_name)
    
    # Удаляем строки с отсутствующими или нераспознанными регионами
    df = df.dropna(subset=[region_column])
    
    # Устанавливаем регион как индекс и сортируем
    return df.set_index(region_column).sort_index()
```

> **poverty_socdem_2017-2020.xls** — распределение малоимущего населения по социально-демографическим группам (дети, трудящиеся, пенсионеры) за 2017–2020 гг., по регионам.

В данных пристутсвуют пропуски и нулевые значения. Соответсвенно их необходимо обработать, так как мера измерения - проценты % ,соответсвенно, в сумме по всем группам должно быть 100% (удалим неинформативный приизнак "Все население").

`Общий вид уравнения:`
$$r = x + y + z$$
,где $r$ (100% населения региона)
 $x$ (процентов детей в возрасте до 16 лет)
 $y$ (процентов населения старше трудоспособного возраста)
 $z$ (процентов населения трудоспособного возраста)

Пропуски обнаруженны в 2019 году и 2020 году.

Заполнение данных: (под пропуском будем подразумевать null или 0)
-   если пропусков >= 2, то заполняем полностью строку прошлогодними данными 

`В результате будем использовать poverty_socdem_2020.`


```python
# Функция для чтения файлов
def poverty_socdem_func(path):
    data = pd.read_excel(path, header=2)
    data.rename({
        'Unnamed: 0': 'region',
        'Все население': 'all',
        'Дети в возрасте до 16 лет': 'children',
        'Население старше трудоспособного возраста': 'pensioners',
        'Население трудоспособного возраста': 'labouring'},
        axis=1, inplace=True)
    
    data.drop('all', axis=1, inplace=True) # удалим неинформативный признак
    names_to_drop = [
        '        Архангельская область',
        '        Тюменская область'
    ]
    
    data = process_region_data(data, standard_names, names_to_drop)
    return data

poverty_socdem_2017 = poverty_socdem_func('social_russia_data/poverty_socdem_2017.xls')
poverty_socdem_2018 = poverty_socdem_func('social_russia_data/poverty_socdem_2018.xls')
poverty_socdem_2019 = poverty_socdem_func('social_russia_data/poverty_socdem_2019.xls')
poverty_socdem_2020 = poverty_socdem_func('social_russia_data/poverty_socdem_2020.xls')

def missing_socdem(data):
    # Считаем количество пропусков и нулей в каждой строке
    missing_count = (data[['children', 'pensioners', 'labouring']].isnull() | 
                    (data[['children', 'pensioners', 'labouring']] == 0)).sum(axis=1)
    
    # исключаем, когда нет пропусков
    mask = (missing_count == 0) | (missing_count == 1)
    missing_count = missing_count[~mask] 
    
    return missing_count

# cводная таблица количетсва пропусков для 2019
missing_count = missing_socdem(poverty_socdem_2019) 

# заполняем пропущенные значения 
mask = poverty_socdem_2019.index.isin(missing_count.index)
poverty_socdem_2019[mask] = poverty_socdem_2018[mask]

# cводная таблица количетсва пропусков для 2020
missing_count = missing_socdem(poverty_socdem_2020) 

# заполняем пропущенные значения 
mask = poverty_socdem_2020.index.isin(missing_count.index)
poverty_socdem_2020[mask] = poverty_socdem_2019[mask]

display(poverty_socdem_2019.describe().round())
display(poverty_socdem_2020.describe().round())
```

> **disabled_total_by_age_2017_2022.csv** — число людей с инвалидностью по регионам, по месяцам, по возрастным группам.

Как всем известно чаще всего человек склонен к инвалидности в пенсионном возрасте, данные это подтверждают. Основную долю инвалидов занимиют люди за 60+. И тут возникает вопрос: "Стоит ли учитывать их в данных?" Я думаю нет, их состояние зачастую связанно с возрастом, поэтому удаляем признак `60_` и создаем признак `18_60`.

Будем использовать данные по концу каждого года то есть по 1 числа декабря, если таких данных нет, то берем последний который есть.

```python
disabled_by_age = pd.read_csv('social_russia_data/disabled_total_by_age_2017_2022.csv')

disabled_by_age = process_region_data(disabled_by_age, standard_names)


disabled_by_age.date = pd.to_datetime(disabled_by_age.date)
# создадим признак год
disabled_by_age['year'] = disabled_by_age['date'].dt.year
# создадим признак месяц
disabled_by_age['month'] = disabled_by_age.date.dt.month 
# число людей с инвалидностью с 18 по 60 лет 
disabled_by_age['18_60'] = disabled_by_age['total'] - disabled_by_age['60_']

disabled_by_age.drop(['total', '18_30',	'31_40', '41_50', '51_60', '60_', 'date'],
                     axis=1,
                     inplace=True)

mask_month = disabled_by_age.month == 12 # месяц декабрь
mask_year = disabled_by_age.year == 2022
mask_month_2022 = disabled_by_age.month == 4

disabled_by_age = disabled_by_age[mask_month|(mask_year & mask_month_2022)]
# сводная итоговая таблица 
disabled_by_age = pd.pivot_table(
    data=disabled_by_age,
    columns='year',
    values='18_60',
    index=disabled_by_age.index
)
display(disabled_by_age.describe().round(2))

```

### **1.5 Жилье и условия жизни**

> **housing_2020** — характеристика жилищных условий домохозяйств, оценка домохозяйствами состояния занимаемого ими жилого помещения, обследование 2020 года.

-   Таблица 1 Характеристика жилищных условий домохозяйств, % от всех домохозяйств
-   Таблица 2 Оценка домохозяйствами состояния занимаемого ими жилого помещения, % от всех домохозяйств

```python
def readout_housing(path, sheet, choice_cols=None):
    data = pd.read_excel(path, sheet_name=sheet, usecols=choice_cols)
    
    data.replace('...', 0, inplace=True) # заполняем пропуски
    data = data.infer_objects(copy=False) # определяем типы данных заново 
    
    data = process_region_data(data, standard_names, region_column='Регион')
    return data

# лист 1
housing_cond = readout_housing('social_russia_data/housing_2020.xlsx',
                               sheet=0,
                               choice_cols='A, E, H')

housing_cond.columns.values[0] = 'домохозяйства с большой стесненностью'

# лист 2
housing_intent = readout_housing('social_russia_data/housing_2020.xlsx',
                                 sheet=1,
                                 choice_cols='A, C:F, H, I')

housing_intent.columns.values[0] = 'улучшить условия'
housing_intent.columns.values[1] = 'стесненность'
housing_intent.columns.values[2] = 'плохие условия'
housing_intent.columns.values[3] = 'стесненность и плохие условия'
housing_intent.columns.values[4] = 'очередь'
housing_intent.columns.values[5] = 'снос дома'

# домохозяйства, которые хотят улучшить свои жилищные условия и рассчитывающих
# на снос или очередь (отсутсвие средств)
housing_intent['отсутствие средств на улучшение'] = housing_intent[['очередь', 'снос дома']].sum(axis=1)

# из тех кто хочет улучшить условия, но не жалуется на текущие условия
housing_intent['улучшить без жалоб'] = (
    housing_intent['улучшить условия'] - 
    ((housing_intent['стесненность'] - housing_intent['стесненность и плохие условия']) +
    (housing_intent['плохие условия'] - housing_intent['стесненность и плохие условия']))
)

# Отбираем только необходимые признаки
housing_intent = housing_intent[['отсутствие средств на улучшение', 'улучшить без жалоб']].copy()
# итоговый датасет 
housing = pd.concat([housing_cond, housing_intent], axis=1)
housing.head()
```

### **1.6 Преступность и социальное неблагополучие**

> **Папка crimes** — сведения о преступлениях, совершённых отдельными категориями лиц за 2016–2022 гг., по месяцам, регионам, категориям лиц, категориям преступлений (требует предобработки).

Будем использовать файл `4-EGS_Razdel_5_062022.xls` - самые актуальные данные о преступлениях среди предоставленных. Данные представленны с января по июнь, то есть за первое полугодие, данных за весь 2022 нет. 

Также файл не отражает полную статистику преступлений за год, она учитывает ***количество предварительно расследованных преступлений в отчетном периоде (из числа находившихся в производстве или зарегистрированных в отчетном периоде) совершенных***

```python
# функция для считывания данных с таблиц о преступлениях
def readout_crimes(path, sheet, choice_cols=None):
    data = pd.read_excel(path, sheet, header=9, usecols=choice_cols)

    data.drop([0, 1], axis=0, inplace= True)
    
    # приведем названия столбцов в нормальное состояние
    data.columns.values[0] = 'region'
    data.columns.values[7] = 'ДомНасилие'
    data.columns.values[8] = 'Иностранцами'
    
    # стандартизируем названия 
    data = process_region_data(data, standard_names)
    
    # преcтупления группой лиц
    data['группой лиц'] = data[[
        'группой лиц',
        'группой лиц по предварительному сговору',
        'организованной группой',
        'преступным сообществом (преступной организацией)'
    ]].sum(axis=1)
    data.drop([
        'группой лиц по предварительному сговору',
        'организованной группой',
        'преступным сообществом (преступной организацией)'], axis=1, inplace=True)
    
    # зарегестрированные преступления
    data['Зарегистрированные преступления'] = data[[
        'несовершеннолетними или при их соучастии',
        'ранее совершавшими преступления'
    ]].sum(axis=1)
    
    data.drop([
        'несовершеннолетними или при их соучастии',
        'ранее совершавшими преступления'
    ], axis=1, inplace=True)
    
    return data


path = 'social_russia_data/crimes/4-EGS_Razdel_5_062022.xls'
choice_cols = 'B:D, F:I, M, P'
# все преступления
crimes_all = readout_crimes(path, sheet=0, choice_cols=choice_cols)

# Особо тяжкие преступления
crimes_severe = readout_crimes(path, sheet=1, choice_cols=choice_cols) 
crimes_severe.columns.values[3] = 'суровые преступления'
crimes_severe = crimes_severe[['суровые преступления']]

# Тяжкие преступления
crimes_heavy = readout_crimes(path, sheet=2, choice_cols=choice_cols)  
crimes_heavy.columns.values[3] = 'тяжкие преступления'
crimes_heavy = crimes_heavy[['тяжкие преступления']]

# Преступления против собственности 
crimes_theft = readout_crimes(path, sheet=12, choice_cols=choice_cols) 
crimes_theft.columns.values[3] = 'преступление собственности'
crimes_theft = crimes_theft[['преступление собственности']]

# преступления по отношению наркотических средств
crimes_drugs = readout_crimes(path, sheet=19, choice_cols=choice_cols)
crimes_drugs.columns.values[3] = 'оборот наркотиков'
crimes_drugs = crimes_drugs[['оборот наркотиков']]

# display('Все преступления:', crimes_all.head(3))
# display('Особо тяжкие преступления:', crimes_severe.head(3))
# display('Тяжкие преступления:', crimes_heavy.head(3))
# display('Преступления по отношению наркотических средств:', crimes_drugs.head(3))

crimes = pd.concat([
    crimes_all,
    crimes_severe,
    crimes_heavy,
    crimes_theft,
    crimes_drugs
], axis=1)

display(crimes.head())
```
